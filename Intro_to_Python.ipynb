### Data Types:

In the previous block, we mixed different data "types" which gave us the error.

Python has a built in set of data types that can be assigned to a variable. When you assign a value to a variable, python automatically decides what the data type should be based on the value that you give it.

Some of the most common data types are: floats, integers, strings, and booleans.


```python
# Float = float. Representation of a real number, which can include decimal places
i_am_a_float = 2.5

# Integer = int. Similar to float, but contains no decimals
i_am_an_integer = 25

# Strings = str. These are text characters always in "quotes", can make numbers as text this way.
# Quotes can be single '' or double "". You just have to use the same type before and after.
i_am_a_string = '25'

# Booleans = bool. These just mean true or false 
my_age_is_25 = True

#Let's test this to make sure this checks out! 
print(type(i_am_a_float)) #the "type" command means I want to know what is the type of variable
print(type(i_am_an_integer))
print(type(i_am_a_string))
print(type(my_age_is_25))
```

```python
#Since I know my data now, I can plot it up to see what it looks like

#both plots below show a different way of calling each column, but will show the same plot

plt.scatter(df.pH, df.Bicarbonate, label='Dot Plot Method')
plt.scatter(df['pH'], df['Bicarbonate'], s=10, label='Index Plot Method') #changed size to show markers in same position

plt.xlabel('pH')
plt.ylabel('Bicarbonate')
plt.legend()
```

```python

# Let's add a trend line to our plot above

xdata = df.pH
ydata = df.Bicarbonate

from scipy import stats

#perform the linear regression
res = stats.linregress(xdata, ydata)

#check the r-squared value
print(f"R-squared: {res.rvalue**2:.6f}")



#plot the original graph 
plt.scatter(df.pH, df.Bicarbonate, label='Original Water Data')

plt.xlabel('pH')
plt.ylabel('Bicarbonate')

#plot the regeression line and use red as color 
plt.plot(xdata, res.intercept + res.slope*xdata, 'r', label='Fitted Line')
plt.legend()
```

```python
#Let's see how a non-linear fit would work if I wanted to do this

from scipy.optimize import curve_fit #import our package for the curve fit
```

Defining a Function to use for a Curve Fit

```python
#to get a curve fit, we need to define our function
#we do this by using the *def* command and using the structure below

def func(x, a, b, c): #these values are general. and now we can import any values we want inside them
  return a * np.exp(-b * x) + c
```

```python
#Here, I'm just going to make some random data
xdata2 = np.linspace(0, 4, 50)
y = func(xdata2, 2.5, 1.3, 0.5)
rng = np.random.default_rng()
y_noise = 0.2 * rng.normal(size=xdata2.size)
ydata2 = y + y_noise
plt.plot(xdata2, ydata2, color='blue', label='random data')
plt.legend()
```

```python
#Now, we want to fit our original data using our new function
popt, pcov = curve_fit(func, xdata2, ydata2) #if you get a runtime error, use 'maxfev=5000'
plt.plot(xdata2, func(xdata2, *popt), 'r-',
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt)) #label pulls values from the curvefit function to find slope and intercepts
plt.plot(xdata2, ydata2, color='blue', label='random data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
