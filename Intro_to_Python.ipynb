```python
#Let's plot up some of our data. How about SpC vs Discharge


plt.scatter(water_quality.Discharge, water_quality.SpC)

plt.xlabel('Discharge (L/s)')
plt.ylabel('SpC $\mu$S/cm');
```

```python
# Let's add a trend line to our plot above

xdata = water_quality.Discharge
ydata = water_quality.SpC

from scipy.stats import linregress

#perform the linear regression
slope, intercept, rvalue, pvalue, stderr = linregress(xdata, ydata)

#plot the original graph 
plt.scatter(water_quality.Discharge, water_quality.SpC, label='Original Water Data')

plt.xlabel('Discharge (L/s)')
plt.ylabel('SpC $\mu$S/cm');

#calculate the r-squared value
R2 = rvalue**2

# We can output this on the plot using plt.text()
# First, convert to a string for display. 
R2_str = '$R^2$=' + str(R2)[:4] 
# Grab only first 4 characters [:4]. You can index a string
# similarly to how a list or array.
# $$ allows you to enter specialized math notation, which is how we get the exponent.
# Think of $$ as 'math quotes."

# Provide x location, y location, and string to print.
plt.text(27, 290, R2_str)

#plot the regeression line and use red as color 
plt.plot(xdata, intercept + slope*xdata, 'r', label='Fitted Line')
plt.legend(); 
#Note: A semi-colon at the end suppresses any print output from the last function
# Sometimes a function will print out some junk we don't need to see.
```

```python
# Other than just straight lines, we could also fit any kind of function we want using curve_fit()

from scipy.optimize import curve_fit #import curve_fit from the scipy package
```

Defining a Function to use for a Curve Fit

```python
#to get a curve fit, we need to define our function
#we do this by using the *def* command and using the structure below

def func(x, a, b, c): #these values are general. and now we can import any values we want inside them
    y = a * np.exp(-b * x) + c
    return y # This is the value that the function gives back when it's done.

# This function has to have the x variable as the first argument, followed by the fitting parameters.
# That said, the x variable doesn't have to be called x.
```

```python
#Here, I'm just going to make some random data
xdata2 = np.linspace(0, 4, 50)
y = func(xdata2, 2.5, 1.3, 0.5)
rng = np.random.default_rng()
y_noise = 0.2 * rng.normal(size=xdata2.size)
ydata2 = y + y_noise
plt.plot(xdata2, ydata2, color='blue', label='random data')
plt.legend()
```

```python
# Now, we want to fit our original data using our new function
# To run curve_fit(), we first provide the fitting function name,
# Then we provide the x and y values of the data, followed by a guess.
# The guess is optional, but it guesses 1 for all of them, if we don't
# provide a guess. Sometimes this will fail (see CO2 notebook for an example).
# Here, we'll just provide the default guess.
p_guess = [1,1,1]

popt, pcov = curve_fit(func, xdata2, ydata2, p_guess) 
# popt contains our optimized parameters for the function, a, b, and c, in the order they appear
# in the function call. We can pull them out like this.
a_opt = popt[0]
b_opt = popt[1]
c_opt = popt[2]

plt.plot(xdata2, func(xdata2, a_opt, b_opt, c_opt), 'r-',label='fit') 
plt.plot(xdata2, ydata2, color='blue', label='random data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

print('a =', a_opt)
print('b =', b_opt)
print('c =', c_opt)
```
